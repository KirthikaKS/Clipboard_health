# Feature Engineering

```python
# Create a Contractor Dependency Ratio
df_staffing["Total_Staffing_Hours"] = df_staffing["Hrs_RNDON_emp"] + df_staffing["Hrs_RNDON_ctr"] + df_staffing["Hrs_LPN_emp"] + df_staffing["Hrs_LPN_ctr"] + df_staffing["Hrs_CNA_emp"] + df_staffing["Hrs_CNA_ctr"]
df_staffing["Contractor_Ratio"] = (df_staffing["Hrs_RNDON_ctr"] + df_staffing["Hrs_LPN_ctr"] + df_staffing["Hrs_CNA_ctr"]) / df_staffing["Total_Staffing_Hours"]
```

```python
# Create a Contractor Dependency Ratio
df_staffing["CNA_Hours_Per_Resident"] = df_staffing["Hrs_CNA"] / df_staffing["MDScensus"]
df_staffing["LPN_Hours_Per_Resident"] = df_staffing["Hrs_LPN"] / df_staffing["MDScensus"]
df_staffing["RN_Hours_Per_Resident"] = df_staffing["Hrs_RNDON"] / df_staffing["MDScensus"]
```

# Data Analysis & Visualizations

## Employee vs. Contractor Staffing


```python
# Summarizing total hours worked by employees vs contractors
staff_type_totals = df_staffing[["Hrs_RNDON_emp", "Hrs_RNDON_ctr", "Hrs_LPN_emp", "Hrs_LPN_ctr", "Hrs_CNA_emp", "Hrs_CNA_ctr"]].sum()

plt.figure(figsize=(10,6))
staff_type_totals.plot(kind='bar', color=['blue', 'orange', 'blue', 'orange', 'blue', 'orange'])
plt.title("Total Hours: Employees vs Contractors")
plt.ylabel("Total Hours Worked")
plt.xticks(rotation=45)
plt.show()

```

## State-Level Contractor Dependence

```python
# Group by state and calculate contractor reliance
statewise_staffing = df_staffing.groupby("STATE")[["Hrs_RNDON_ctr", "Hrs_LPN_ctr", "Hrs_CNA_ctr"]].sum()
statewise_staffing["Total_Contractor_Hours"] = statewise_staffing.sum(axis=1)
statewise_staffing = statewise_staffing.sort_values("Total_Contractor_Hours", ascending=False)

# Bar chart of contractor reliance per state
plt.figure(figsize=(12, 6))
statewise_staffing["Total_Contractor_Hours"].plot(kind='bar', color="darkblue")
plt.title("Total Contractor Hours by State")
plt.ylabel("Total Hours")
plt.xticks(rotation=90)
plt.show()
```

## Facilities with High Contractor Reliance

```python
# Find facilities with highest contractor dependency
facility_contractor_reliance = df_staffing.groupby(["PROVNUM", "PROVNAME", "STATE"])[["Contractor_Ratio"]].mean()
facility_contractor_reliance = facility_contractor_reliance.sort_values("Contractor_Ratio", ascending=False)

# Display top facilities
print("\nTop Facilities by Contractor Dependence:")
print(facility_contractor_reliance.head(10))
```

## Relationship Between Resident Census & Staffing

```python
# Scatter plot to show relationship between resident census and total CNA hours
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df_staffing["MDScensus"], y=df_staffing["Hrs_CNA_emp"] + df_staffing["Hrs_CNA_ctr"], alpha=0.5)
plt.title("Resident Census vs. CNA Staffing Hours")
plt.xlabel("Number of Residents (MDScensus)")
plt.ylabel("Total CNA Staffing Hours")
plt.show()
```

```python
# Visualization: Understaffed Facilities by State
plt.figure(figsize=(12,6))
states_understaffed = df_staffing.groupby("STATE")["Understaffed_Flag"].sum().sort_values(ascending=False)
states_understaffed.plot(kind='bar', color='red')
plt.title("Number of Understaffed Facilities by State")
plt.ylabel("Count of Understaffed Facilities")
plt.xticks(rotation=90)
plt.show()
```

### 2.Focus on Understaffed High-Census Facilities

#### Why?

Nursing homes with high residents but low staffing are at risk of compliance issues.
These homes urgently need contractor staffing.

#### Action Plan:

Find facilities where CNA hours per resident < state median.
Offer CNA contractors to improve compliance.

#### Revenue Impact: 

Fast sales conversion due to urgent demand.

```python
state_median_cna = df_staffing.groupby("STATE")["CNA_Hours_Per_Resident"].median()
df_staffing["Understaffed_Flag"] = df_staffing.apply(lambda x: x["CNA_Hours_Per_Resident"] < state_median_cna[x["STATE"]], axis=1)

df_understaffed = df_staffing[df_staffing["Understaffed_Flag"]]
```

Maximizes profit from urgent staffing needs.
